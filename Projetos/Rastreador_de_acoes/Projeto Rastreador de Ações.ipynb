```python
import yfinance as yf
import matplotlib.pyplot as mplt
```

---
3. Definir as ações
> Escolha as ações que deseja rastrear e defina seus símbolos como uma lista. Se for ação brasileira, adicione ".SA" ao final. Lembrando que cada ação terá o valor da moeda local.
---

```python
acoes = ['AAPL', 'MSFT', 'PETR4.SA', 'BBDC4.SA']
```

---
4. Importar os dados
> Usando a biblioteca yfinance para obter os dados das ações.
---

```python
dados_acoes = yf.download(acoes,
                          period = '1d',
                          start = '2022-01-01')
```

---
5. Criando o primeiro gráfico
> Vamos plotar os preços de fechamento das ações no gráfico.
---

```python
dados_acoes['Close'].plot()
mplt.xlabel('Data')
mplt.ylabel('Preço de Fechamento')
mplt.title('Rastreador de ações')
mplt.legend(acoes, fontsize = 8)
mplt.show()
```

---
6. Normalizando os preços
> Como cada ativo parte de um valor inicial diferente, vamos normalizar para que todos partam do mesmo ponto, a fim de obter uma comparação mais **precisa**.
---

```python
dados_normalizados = dados_acoes['Close'] / dados_acoes['Close'].iloc[0]
dados_normalizados.plot()
mplt.xlabel('Data')
mplt.ylabel('Preço Normalizado')
mplt.title('Rastreador de ações')
mplt.legend(acoes, fontsize = 8)
mplt.show()
```
